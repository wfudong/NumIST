# 01 - One Soliton

This notebook validates the zero-background Prins-Wahls Crum implementation on the one-soliton case

$$q(x)=2\kappa^2\operatorname{sech}^2(\kappa x).$$

```julia
using Pkg
Pkg.activate("..")

using Revise
includet("../src/KdVCrumProject.jl")
using .KdVCrumProject

using Plots, LaTeXStrings
```

For a single eigenvalue, the target norming constant `beta=1` centers the soliton at the origin.

```julia
x = collect(range(-12.0, 12.0; length=2001))
kappa = [1.4]
beta = [1.0]

q = crum_transform(x, kappa, beta)
q_exact = exact_one_soliton(x, kappa[1]; beta=beta[1])

relerr = relative_l2_error(q, q_exact)
maxerr = maximum(abs.(q .- q_exact))
E1, E2 = conserved_errors(x, q, kappa)

@show relerr
@show maxerr
@show minimum(q)
@show maximum(q)
@show E1 E2;
```

```julia
plot_comparison(x, q, q_exact; title="One-soliton validation")
```

A shifted one-soliton can be produced directly with `shifts=[s]`. For `beta=1`, the center is at `x=s`.

```julia
shift = 1.2
q_shifted = crum_transform(x, kappa, beta; shifts=[shift])
q_shifted_exact = exact_one_soliton(x, kappa[1]; beta=beta[1], shift=shift)

@show relative_l2_error(q_shifted, q_shifted_exact)
plot_comparison(x, q_shifted, q_shifted_exact; title="Shifted one soliton")
```

The repeated-Darboux reference agrees in this one-eigenvalue case; differences show up for multi-soliton spectra.

```julia
q_naive = naive_darboux_chain(x, kappa, beta)
@show relative_l2_error(q_naive, q_exact)
plot_comparison(x, q_naive, q_exact; title="Naive one-step Darboux check")
```
